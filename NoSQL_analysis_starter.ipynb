### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene':{'$eq':20}}
fields = {}
limit = 1
result = list(establishments.find(query, fields))
# Use count_documents to display the number of documents in the result
resultcount = establishments.count_documents(query)
print(f'There are {resultcount} establishments with a hygiene score of 20.')
# Display the first document in the results using pprint
resultviewlimit = list(establishments.find(query, fields).limit(limit))
pprint(resultviewlimit)
```

```python
# Convert the result to a Pandas DataFrame
result_df = pd.DataFrame(result)
# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(result_df))
# Display the first 10 rows of the DataFrame
result_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
querya = {"AddressLine3":{'$regex':"London"},
        'RatingValue': {'$gt':4}}
resulta = list(establishments.find(querya, fields))
# Use count_documents to display the number of documents in the result
resultcounta = establishments.count_documents(querya)
print(f'There are {resultcounta} establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.')
# Display the first document in the results using pprint
resultviewlimita = list(establishments.find(querya, fields).limit(limit))
pprint(resultviewlimita)
```

```python
# Convert the result to a Pandas DataFrame
result_dfa = pd.DataFrame(resulta)
# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(result_dfa))
# Display the first 10 rows of the DataFrame
result_dfa.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
#2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipeline = [
    {"$match": {'scores.Hygiene': 0}},
    {"$group": {'_id' : '$LocalAuthorityName', "count":{ '$sum':1}}},
    {'$sort':{'count':-1}}
]
pipelinelimten = [
    {"$match": {'scores.Hygiene': 0}},
    {"$group": {'_id' : '$LocalAuthorityName', "count":{ '$sum':1}}},
    {'$sort':{'count':-1}},
    {'$limit':10}
]
pipeline_count = [
    {"$match": {'scores.Hygiene': 0}},
    {"$group": {'_id' : '$LocalAuthorityName', "count":{ '$sum':1}}},
    {'$sort':{'count':-1}},
    { '$count': "Total" }]
resultsgb = establishments.aggregate(pipeline)
resultscount = establishments.aggregate(pipeline_count)
resultslimten = establishments.aggregate(pipelinelimten)
# Print the number of documents in the result
pprint(list(resultscount))
# Print the first 10 results
pprint(list(resultslimten))
```

```python
# Convert the result to a Pandas DataFrame
result_dfg = pd.DataFrame(resultsgb)
# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(result_dfg))
# Display the first 10 rows of the DataFrame
result_dfg.head(10)
```
